# 5.1 培养“品味”，用 AI 为业务提效

## 🚄 前言

你已经学习了从 RAG 到 Agent 的相关技术，一定也在思考着如何把 AI 用到自己的业务中。从客服答疑、合同审核、内容翻译、代码辅助，到数据分析，每一个方向似乎都很有价值，每个业务需求都很紧急。从哪里切入既能快速落地，又能满足业务需要，产生更大的业务价值？你需要的不只是搭系统的能力，还需要选方向、定标准、验效果的品味。

## 🍁 课程目标

**你将学到：**
- 掌握在业务中应用 AI 的四个阶段即 RIDE 方法论
- 学会判断一个业务场景是否适合用 AI 解决，理解如何推动 AI 在业务落地
- 理解有意识地培养自己的“品味”，提升自己的 AI 技能

1. 重构生产关系：让业务主导 AI 建设你一个人开发答疑机器人，不需要和任何人协调。在真实团队中，事情没这么简单。AI 是新的生产力，但传统企业原有的组织结构、协作方式、责任归属，也就是生产关系，还没有适应 AI 时代。当你试图把答疑机器人推广到公司，你可能会在许多环节发现这种不适应：业务觉得 AI 是技术的事，技术觉得业务不提需求，没人对最终效果负责。这就是旧的生产关系在阻碍新生产力的落地。所以 AI 落地的第一步，不是写代码，而是重构生产关系。核心原则只有一个：业务主导。在 Agent 时代，这个趋势更加明显。业务专家最理解问题、最清楚目标，也越来越有能力自主闭环。技术团队的角色从“主导开发”转向“支撑赋能”——在很多场景下，前端、后端、算法的边界已经模糊，技术是底座，不是主角。回到答疑机器人。它上线后应该“汇报”给谁？你可以这样做：把答疑机器人这类 AI 数字员工汇报给业务部门，而你作为技术团队退到“供应商”的位置，业务部门和技术团队联合培养数字员工。业务团队主导目标、主导评判、主导迭代方向——这才是真正的生产关系重构。要让这种新的生产关系跑起来，有两件事要先做到位。第一，做 AI 通识教育。 业务团队要主导 AI，首先得理解 AI 能做什么、不能做什么。在企业落地 AI 的过程中你会发现：产品经理、算法工程师、业务经理对大模型的理解完全不同。大家都在谈同一个概念，但理解的内容和现实完全不一样。如果没有做过深入的落地工作，很难体会到那种无力感。为了解决这个问题，企业管理者会自然想到做培训，实现“书同文、车同轨”——要求全员参加 AI 认证培训。ACA 面向非技术人员，ACP 面向技术人员。没错，你正在学的这门课，就是这个行动的产物。我们称之为“AI 时代的通识教育”。目前，整个阿里巴巴集团都在用这套方法做 AI 转型的基础教育。一旦通过通识教育统一了认知，沟通效率会显著提升，业务团队也才真正有能力主导 AI 的方向。第二，确立正确的评判标准。 业务主导意味着由业务来定义什么叫“好”。你可能会听到这样的质疑：“大模型有幻觉，做不到 100% 正确。”但实践告诉我们，人也有“幻觉”——老员工回答新人问题时也会出错，只是企业没注意到。评判标准不是“AI 是否完美”，而是“AI 是否比人做得更好”。这就是一个重要的原则：“和人比，不和神比”——AI 的效率和效果都比人工好，就可以上岗。要求 100% 正确，就是在和“神”比。如果不先解决生产关系的问题，业务部门就会拿“100% 正确”来要求 AI——这就是旧的评判标准在阻碍新生产力。在 AI 时代重构生产关系，让业务主导、技术支撑、全员对齐认知、用业务目标驱动评判，即 Reorganize。不解决这个问题，后面每一步都会遇到阻碍。2. 如何选择合适的 AI 落地点你学习整个课程开发了一款新人答疑机器人。本课程为什么选这个场景，而不是合同审核、内容翻译或者数据分析？因为它同时满足三个特征：Language 类：新员工提问是自然语言，回答也是自然语言。输入输出都围绕文本展开，天然适合大模型。重复多：新员工往往频繁遇到诸如“怎么请假”、“用什么项目管理工具”的共性问题，流程化且可预期。业务压力大：回答新人问题不是老员工的核心职责，但不答又影响新人效率——这是典型的“多任务岗位的繁琐子任务”。有些单任务岗位（如翻译）可以直接用 AI 线性替换产能；但像产品经理、财务、法务这类多任务岗位，更有效的做法是把其中重复繁琐的部分卸载给 AI，让人聚焦在真正需要判断和创造的工作上。能同时满足这三个特征，这种场景的优先级就很高。还有一个容易忽略的前提：公司是否已经有组织好的文档库。这是答疑机器人这个项目的“蛋糕坯”。AI 是上面的“樱桃”——锦上添花。如果连文档都没有，AI 也无从检索。蛋糕坯没做好，放再多樱桃也不行。找出适合 AI 切入的业务应用点，即 Identify，就通过这三点：Language 类、重复多、业务压力大。再检查基础设施是否 ready，就可以确定是否是最佳的业务切入点。理顺了组织关系、确定好场景后，接下来的问题是：怎么判断做得好不好？这比写代码更难。3 怎么判断做得好不好？你在 2.4 用 RAGAS 建了评测体系，在 3.6 用评测驱动 Agent 开发。评测驱动是这门课贯穿始终的主线。但系统上线后，“做得好不好”的定义会发生变化。AI 产品和传统互联网产品有很多相似之处——都有交互、有体验、有用户旅程。但 AI 产品有一个核心区别：准确率。准确率不够好，用户根本无法使用，数字员工也不可能真正上岗。除了准确率，还有响应时效性（比如电销场景中，延迟太高客户会觉得在和机器说话）和安全合规等非功能性指标。但准确率是第一核心，整个项目组都必须盯住它。这里有一个常见的陷阱。你可能看到过类似“用这一招准确率提升到 95%”的文章，但这些往往有一个前提：在某个狭窄的小场景下。回到答疑机器人，50 条常见问题的准确率可能很高，但真实用户的问题千奇百怪，面对海量真实问题时，准确率未必能维持。只有准确率够吗？不够。如果只盯产品指标，大概率会掉到坑里。移动互联网时代的运营基本功不能丢：层级指标示例回答的问题产品指标准确率、响应时效性、安全合规AI 能力本身够不够好运营指标DAU、留存率、渗透率、提问数用户真的在用吗？用了还回来吗？技术指标RAGAS 分数、检索召回率哪个技术环节需要优化其中留存率最关键。新员工今天用了答疑机器人，下周还愿意继续用，说明它真的解决了问题。如果只用一次就不回来，那可能不是算法需要调优，而是问题定义错了。运营指标的作用是防止“自嗨”——什么叫自嗨？你觉得你的答疑机器人，从产品、工程和算法的角度来看“指标很好”，结果公司的新员工们根本不用。不紧盯运营指标，研发团队很容易陷入这种自我满足。这就引出了“品味”最核心的含义：品味 = 知道如何评测。AI 时代最难的不是搭系统，而是定义什么叫“好”。你在 2.4 构建评测数据集时已经体验过——Ground Truth 需要业务专家提供，工程师猜不出来。到了 3.6 评测驱动开发，这个感受会更深：评测设计本身值得你做最大的投入。从产品指标到运营指标再到技术指标，精准地定义度量体系，即 Define。从项目启动就想清楚这三层指标怎么关联，才能避免自嗨，确保 AI 真正产生业务价值。理顺了组织关系、确定好场景、定义了度量体系，你开始动手了。从 C2 到 C3，这条路径本身就包含了几个重要的模式。4. 动手：从翻译模式到 Agent 模式翻译模式： 回顾 C2 阶段，你搭的 RAG 系统做一件事：把用户的问题翻译成知识检索，再把检索结果翻译成回答。输入是自然语言，输出也是自然语言，任务边界清晰固定。这本质上是一种翻译操作。为什么大模型特别擅长翻译？因为 Transformer——大模型背后的核心算法——最早就是 Google 为翻译任务开发的。翻译可以分为狭义和广义两种。狭义翻译是语言之间的转换，比如中译英、英译中。广义翻译的范围更大：自然语言翻译成 SQL、翻译成代码、把一篇论文“翻译”成通俗表述，都属于广义翻译。Transformer 对这两类翻译都特别擅长，所以翻译模式是 AI 落地中最容易摘到的“低垂果实”——落地快、容易评测、风险低。但翻译模式有一个前提：原有系统必须 ready。以 Chat BI 为例：它的本质是“自然语言→SQL→执行结果→自然语言”的双重翻译。但如果企业的数据口径混乱、业务逻辑不统一，连资深分析师手写 SQL 都经常返工；人都翻译不出来，AI 也不可能翻译成功。这就是第 2 节讲的“蛋糕坯”问题：蛋糕坯没做好，放再多樱桃也不行。从翻译者到助手： 接着从课时 3.0 开始，同事提出更多要求：“帮我搜最新行业报告”“帮我更新项目进度”“帮我审一下这个文档”。机器人只会回答，不会动手。它是一个顾问，不是一个助手。于是你在 C3 给它加上了新能力：工具调用（3.1）让它能搜索和操作；规划与反思（3.2）让它先想再做；多 Agent 协作（3.3）让多个专家协同工作；记忆（3.4）让它积累经验；技能（3.5）让能力可复用；评测驱动开发（3.6）让质量有保障。系统从“翻译者”变成了“助手”——这就是Agent 模式。Agent 模式有一个核心原则：“始于用户意图，终于意图满足”。如果你的 Agent 不是从用户意图出发，也不以是否满足用户意图来度量，它很难成功。怎么落地这个原则？靠技术团队单打独斗是不行的。谁最理解用户意图？业务专家。谁能判断知识是否完备？业务专家。谁来定义评测标准？还是业务专家。这就是第 1 节讲 Reorganize 的意义：只有业务真正主导了，下面四步才走得通：建立意图空间。 你的用户会提出哪些类型的需求？以答疑机器人为例，新员工的问题可以从 HR 工单中分析——请假流程、工具使用、报销规则、团队介绍……把这些意图穷举出来、分好类，就是意图空间。它决定了你的 Agent 要覆盖多大的范围。做好知识工程。 基于意图空间，逐类检查知识储备是否完备。比如“报销规则”这类意图，对应的财务制度文档是否在知识库里？如果某类意图对应的知识完全缺失，那这类问题不可能被正确回答。带着意图做评测。 意图和知识都清楚了，才能真正开展评测。这和你在 3.6 做的评测驱动开发一脉相承——但维度从“单个问答对是否正确”扩展到了“用户意图是否被满足”。E2E 归因。 当一个 Bad case 出现时，你需要定位问题出在哪个环节。比如新员工问“怎么申请出差补贴”，机器人答错了——是检索没找到相关文档？还是文档里根本没有出差补贴的内容？还是检索到了但模型理解错了？每一个 Bad case 都应该归因到具体的工程环节，才能聚类改进。意图空间越大，所需的知识准备、工程开发和迭代投入就越大。所以开发 Agent 应用，首先应该从满足较小的、明确的意图集合开始建设，逐步扩大和跟进意图空间的规模和变化，不断完善知识空间。模式核心特征对应课程翻译模式边界清晰、输入输出固定、原有系统必须 readyC2（2.1 - 2.5）Agent 模式始于意图、终于满足、需要意图空间和知识工程C3（3.1 - 3.8）大多数项目应该从翻译模式起步，快速评测和迭代，小步快跑树立标杆，也为后续向 Agent 模式演进积累数据基础。大部分问题出在数据层面： 无论是翻译模式还是 Agent 模式，如果你具备了度量能力，会发现一个规律：绝大多数 Bad case 的根因指向数据。比如文档质量不高、异质化数据多和系统 API 不完善。你在 2.5 优化 RAG 准确率时就体验过：最大的改进来自文档解析和知识库质量的提升，而不是模型升级。你可能会问：什么时候需要引入模型训练？答案是：当你已经用好了基模 API、积累了充分的评测数据、做了 E2E 归因迭代，发现基模确实无法解决某个具体问题时：比如实时性要求极高的场景，可以用小模型训练来解决，这就是“大小结合”。但如果评测能力还不 ready，投入在训练上的每一分钱都是浪费。这也是课程不涉及模型训练的原因：绝大多数场景，用好基模就够了。从这些经验中可以提炼出一个结构：如果把 AI 落地画成一个金字塔，底座是数据与评测——业务数据、业务 API、评测能力，这部分的投入必须充足。中间层是工程应用算法——RAG、微调、提示词优化，这些是必要条件，但不是充分条件。顶部是业务目标拉动——只有用真正的业务痛点去牵引，才能动员领域专家参与评测，才能让项目持续获得资源支持。RIDE 把这个阶段叫 Execute——落地实施与持续迭代。📝 5. 带着这套方法开始你的下一个项目回顾一下这门课的四个阶段。你先解决了一个组织问题：谁来主导 AI 建设？答案是业务主导、技术支撑，通过通识教育统一认知，用“和人比，不和神比”校准评判标准——这叫 Reorganize。课程里你一个人开发，这一步很简单；但在真实团队中，它往往是耗时最长、也最容易被跳过的环节。然后你选了一个合适的切入点。新人答疑满足三个特征：Language 类、重复多、业务压力大，而且公司已有文档库作为蛋糕坯——这叫 Identify。场景选好了，你定义了什么叫“好”：准确率是核心产品指标，留存率防止自嗨，RAGAS 分数定位技术瓶颈。评测设计本身值得最大的投入——这叫 Define。最后你动手了。从 C2 的翻译模式起步，快速验证；到 C3 演进为 Agent 模式，用意图空间驱动知识工程和评测。一路走下来，最大的改进来自数据层面，而不是算法升级——这叫 Execute。这四个阶段连起来就是 RIDE（Reorganize → Identify → Define → Execute），来自业界大模型应用落地项目的实践经验和专家之间的讨论总结。每个阶段的核心都是一个判断——选什么场景、定什么标准、怎么取舍。做出这些判断的能力，就是品味。品味不是读出来的，是在项目中练出来的。下次启动 AI 项目前，问自己四个问题：团队认知一致吗？场景选对了吗？成功标准定了吗？数据和评测底座够吗？不需要长文档，每个问题写一句话就够了。下一节，你将看到更广阔的 AI 应用图景，了解接下来该往哪个方向走。🔥 课后小测验🔍 单选题 5.1.1你的老板让你用大模型做一个 Chat BI 产品，让业务人员用自然语言查询数据。你调研后发现，公司数据仓库里同一个指标在不同部门有三种不同的计算口径，连资深分析师手写 SQL 都经常因为口径问题返工。你的下一步应该是什么❓A. 先选一个口径最清晰的业务线做 MVP，用小范围成功证明价值后再推广B.
